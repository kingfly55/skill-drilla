# Corpus audit

This notebook demonstrates how to audit canonical evidence artifacts for included, excluded, and ambiguous material without reparsing transcripts. It reads only files already materialized under `artifacts/chat-analysis/`.


```python
from pathlib import Path

from skill-drilla.notebooks.loaders import (
    collect_evidence_by_status,
    load_corpus_view,
    load_evidence,
    load_inventory,
    load_parse_diagnostics,
    recurrence_snapshot,
)

REPO_ROOT = Path('.')
ARTIFACT_ROOT = REPO_ROOT / 'artifacts/chat-analysis/'
INVENTORY_PATH = ARTIFACT_ROOT / 'discovery' / 'smoke-a' / 'session_inventory.jsonl'
PARSE_DIAGNOSTICS_PATH = ARTIFACT_ROOT / 'parse' / 'smoke' / 'parse_diagnostics.json'
EVIDENCE_PATH = ARTIFACT_ROOT / 'normalize' / 'smoke' / 'evidence.jsonl'
DEBUG_VIEW_DIR = ARTIFACT_ROOT / 'views' / 'debug_included_and_excluded'
```

```python
inventory = load_inventory(INVENTORY_PATH)
parse_diagnostics = load_parse_diagnostics(PARSE_DIAGNOSTICS_PATH)
evidence_sample = load_evidence(EVIDENCE_PATH, limit=2000)
status_groups = collect_evidence_by_status(evidence_sample)
debug_view = load_corpus_view(DEBUG_VIEW_DIR, limit=25)

{
    'inventory_rows': len(inventory),
    'parsed_files': parse_diagnostics['aggregate']['files'],
    'status_counts': {key: len(value) for key, value in status_groups.items()},
    'view_name': debug_view['view_name'],
    'view_recurrence_basis': debug_view['summary']['recurrence_basis'],
}
```

```python
snapshot = recurrence_snapshot(row['evidence'] for row in debug_view['rows'])
first_rows = [
    {
        'evidence_id': row['evidence']['evidence_id'],
        'semantic_class': row['evidence']['semantic_class'],
        'inclusion_status': row['evidence']['inclusion_status'],
        'source_anchor': row['inspection']['source_anchor'],
    }
    for row in debug_view['rows'][:5]
]

{'snapshot': snapshot, 'sample_rows': first_rows}
```
